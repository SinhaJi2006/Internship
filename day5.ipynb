```python
import matplotlib.pyplot as plot
```

```python
x= [1,2,3,4,5]
y=[10,20,30,40,50]
```

```python
plot.plot(x,y)
plot.title("student hrs vs marks")
plot.xlabel("hrs")
plot.ylabel("marks")
plot.show()
```

```python
subject = ["M1" , "M2" , "PHYSICS", "CHEMISTRY"]
MARKS = ["19" , "18" , "15", "17"]
```

```python
plot.plot(subject,MARKS)
plot.title("student hrs vs marks")
plot.xlabel("subject")
plot.ylabel("MARKS")
plot.show()
```

```python
plot.bar(subject,MARKS)
plot.title("student sub vs marks")
plot.xlabel("subject")
plot.ylabel("MARKS")
plot.show()
```

```python
plot.scatter(subject,MARKS, color='red')
plot.title("student hrs vs marks")
plot.xlabel("subject")
plot.ylabel("MARKS")
plot.show()
```

```python
plot.plot(subject,MARKS , color = 'y')
plot.title("student hrs vs marks")
plot.xlabel("subject")
plot.ylabel("MARKS")
plot.show()
```
